```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x,y)

axes[0][1].plot(x,y)

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y Label')
axes[1][0].set_xlim(2,6)

axes[1][1].plot(x,y)
axes[1][1].set_title('Title')

fig.suptitle('Figure Level', fontsize=12)

fig.subplots_adjust(wspace=0.5, hspace=0.5)
# plt.tight_layout()

fig.savefig('new_subplots.png', bbox_inches='tight')
```

# Legend

```python
x = np.linspace(0, 10, 11)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, label='X vs X')
ax.plot(x, x**2, label='X vs X^2')
ax.legend(loc=(1.1, 0.5))
```

### Me messing around with matplotlib

```python
n = np.linspace(1, 5)
```

```python
from scipy.special import factorial

fig = plt.figure(dpi=125)
ax = fig.add_axes(rect=[0, 0, 1, 1])
ax.set_title("Big-O Complexity")
ax.axhline(label='O(1)')
ax.plot(n, n, label='O(n)')
ax.plot(n, np.log(n), label='O(log n)')
ax.plot(n, n * np.log(n), label='O(n log n)')
ax.plot(n, n**2, label='O(n^2)')
ax.plot(n, 2**n, label='O(2^n)')
ax.plot(n, factorial(n), label='O(n!)')
ax.legend()
fig.savefig('Big-O Chart.png')
```

## Matplotlib Styling

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, color='orangered', lw=5, marker='+',ls='--', ms=20)
lines = ax.plot(x, x+1, color='purple',linewidth=2, marker='o', linestyle='--', markersize=10,
               markerfacecolor='red', markeredgewidth=4, markeredgecolor='orange')
```
